```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

conteo = datasin999.groupby("PROFESION")['GENERO'].value_counts().unstack(fill_value=0)

num_barras = len(conteo)
profesiones = conteo.index.tolist()
generos = conteo.columns.tolist()

fig, ax = plt.subplots(figsize=(14, 7))  # Más ancho para etiquetas

def gradient_rect_vertical(ax, xy, width, height, color1, color2):
    cmap = LinearSegmentedColormap.from_list("", [color1, color2])
    gradient = np.linspace(0, 1, 256).reshape(-1, 1)
    im = ax.imshow(gradient, extent=(xy[0], xy[0]+width, xy[1], xy[1]+height),
                   origin='lower', aspect='auto', cmap=cmap)
    rect = patches.Rectangle(xy, width, height, fill=False, edgecolor='black', linewidth=1.2)
    ax.add_patch(rect)
    return im

bar_width = 0.35
bar_spacing = 0.1
indices = np.arange(num_barras)
colores_base = ['magenta', 'blue']

# Dibujar barras agrupadas con degradado vertical
for i, genero in enumerate(generos):
    for j in range(num_barras):
        valor = conteo.iloc[j, i]
        x = indices[j] + i*(bar_width + bar_spacing)
        y = 0
        gradient_rect_vertical(ax, (x, y), bar_width, valor, colores_base[i], 'white')

# Ajustar ejes
ax.set_xticks(indices + bar_width + bar_spacing / 2)
ax.set_xticklabels(profesiones, rotation=60, ha='right', fontsize=10)
ax.set_ylabel("Cantidad")
ax.set_xlabel("Profesión")
ax.set_title("Distribución de Género por Profesión con Degradado Vertical")
ax.set_xlim(-0.5, num_barras - 0.5 + 2*(bar_width + bar_spacing))
ax.set_ylim(0, conteo.values.max()*1.1)

# Leyenda manual
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='magenta', lw=4, label=generos[0]),
    Line2D([0], [0], color='blue', lw=4, label=generos[1])
]
ax.legend(handles=legend_elements, title='Género')

plt.tight_layout()
plt.show()
```

**Demostrativo.- Análisis de correlación**

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr, kendalltau

# Codificar LIM_INGRESO si es categórico
datamergesin999['LIM_INGRESO_Cod'] = datamergesin999['LIM_INGRESO'].astype('category').cat.codes

x = datamergesin999['LIM_INGRESO_Cod']
y = datamergesin999['Comportamiento general']

# Pearson
pearson_corr, pearson_p = pearsonr(x, y)

# Spearman
spearman_corr, spearman_p = spearmanr(x, y)

# Kendall
kendall_corr, kendall_p = kendalltau(x, y)

print(f"Coeficiente Pearson: {pearson_corr:.3f} (p={pearson_p:.3f})")
print(f"Coeficiente Spearman: {spearman_corr:.3f} (p={spearman_p:.3f})")
print(f"Coeficiente Kendall: {kendall_corr:.3f} (p={kendall_p:.3f})")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Codificar LIM_INGRESO
datamergesin999['LIM_INGRESO_Cod'] = datamergesin999['LIM_INGRESO'].astype('category').cat.codes

x = datamergesin999['LIM_INGRESO_Cod']
y = datamergesin999['Comportamiento general']

plt.figure(figsize=(8,5))

# Añadir jitter
x_jitter = x + np.random.normal(0, 0.1, size=len(x))
y_jitter = y + np.random.normal(0, 0.05, size=len(y))

plt.scatter(x_jitter, y_jitter, alpha=0.4, color='purple')
plt.xlabel('LIM_INGRESO (Codificado)')
plt.ylabel('Comportamiento general (0=Bueno,1=Malo)')
plt.title('Gráfico de dispersión con jitter entre LIM_INGRESOnivel de ingreso y Comportamiento')
plt.grid(True)
plt.show()
```
